```python
import pandas as pd 
import datetime 
import numpy as np
pd.options.display.max_colwidth = 1000
```

```python
n=200
poblaciones = [.5, 1, 1.5]
dates=pd.DataFrame({'Dates':pd.date_range(start='01/01/2020', end='15/12/2022')})
```

```python
dates.sample(n=1).iloc[0]
```

```python
np.random.exponential(size=0)
```

```python
dfs=[]
h=0
for p in poblaciones:
    for i in range(h+1,h+201):
        fechas=[]
        fechas.append(dates.sample(n=1)['Dates'].iloc[0])
        for j in range(1,10):
            time_delta=np.random.exponential(size=1,scale=1/p)*100
            fechas.append(fechas[j-1]+datetime.timedelta(days=time_delta[0]))
        dfs.append(pd.DataFrame({'User_id':[i for a in range(10)],'Dates':fechas,'Poblacion':[p for b in range(10)]}))
    h+=200
```

```python
aux=pd.concat(dfs)
```

```python
aux.reset_index(drop=True)
```

```python
aux=aux.merge(aux.groupby(['User_id'])['Dates'].min(),how='left',on='User_id')
```

```python
aux.head()
```

```python
# aux.assign(min_date_transform = lambda df: df.groupby('User_id').transform(lambda df: df.min())).head()
```

```python
aux.rename({'Dates_x':'Dates','Dates_y':'Min'},inplace=True)
```

```python
aux
```

```python
aux['dif']=aux.apply(lambda x: (x['Dates_x']-x['Dates_y']).days//7,axis=1)
```

```python
aux
```

```python
aux['W1']=aux.apply(lambda x: 1 if x['dif']==1 else 0, axis=1)
```

```python
for i in range(1,13):
    aux[f'W{i}']=aux.apply(lambda x: 1 if x['dif']==i else 0, axis=1)
```

```python
results=aux[['User_id','dif','Poblacion','W1', 'W2', 'W3',
       'W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10', 'W11', 'W12']]
```

```python
results
```

```python
results.drop_duplicates(['User_id','dif'],inplace=True)
```

```python
results=results.groupby(['Poblacion'])[['W1', 'W2', 'W3',
       'W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10', 'W11', 'W12']].sum()
```

```python
final=results.applymap(lambda x: x/2)
```
